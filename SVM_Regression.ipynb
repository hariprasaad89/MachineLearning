```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
dependent=dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="linear")
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
regressor.n_support_
```

```python
regressor.intercept_
```

```python
from sklearn.metrics import r2_score
r2_score=r2_score(y_test,y_pred)
r2_score
#r2_score_rbf=r2_score(y_test,y_pred_rbf)
```

```python
from sklearn.svm import SVR
regressor_rbf=SVR(kernel="rbf")
```

```python
regressor_rbf.fit(x_train,y_train)
```

```python
y_pred_rbf=regressor_rbf.predict(x_test)
```

```python
regressor_rbf.intercept_
```

```python
regressor_rbf.n_support_
```

```python
from sklearn.metrics import r2_score
r2_score_rbf=r2_score(y_test,y_pred_rbf)
r2_score_rbf
```

```python
from sklearn.svm import SVR
regressor_poly=SVR(kernel="poly")
```

```python
regressor_poly.fit(x_train,y_train)
```

```python
y_pred_poly=regressor_poly.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score_poly = r2_score(y_test,y_pred_poly)
r2_score_poly
```

```python
from sklearn.svm import SVR
regressor_sigmoid=SVR(kernel="sigmoid")
regressor_sigmoid.fit(x_train,y_train)
```

```python
y_pred_sigmoid=regressor_sigmoid.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score_sigmoid=r2_score(y_test,y_pred_sigmoid)
r2_score_sigmoid
```

```python
from sklearn.svm import SVR
regressor_precomputed=SVR(kernel="precomputed")

regressor_precomputed.fit(x_train,y_train)
#so precomputed algm will not work for this dataset (only for this dataset only)
```

```python
#Standarization - need to apply this technique to input data file for converting the columns values to standards
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
regressor_sigmoid.fit(x_train_sc,y_train)
regressor_poly.fit(x_train_sc,y_train)
regressor_rbf.fit(x_train_sc,y_train)
regressor.fit(x_train_sc,y_train)
```

```python
y_pred_sc=regressor.predict(x_test_sc)
y_pred_rbf_sc=regressor_rbf.predict(x_test_sc)
y_pred_sigmoid_sc=regressor_sigmoid.predict(x_test_sc)
y_pred_poly_scc=regressor_poly.predict(x_test_sc)
```

```python
from sklearn.metrics import r2_score
r2_score_sc = r2_score(y_test,y_pred_sc)
r2_score_rbf_sc = r2_score(y_test,y_pred_rbf_sc)
r2_score_sigmoid_sc = r2_score(y_test,y_pred_sigmoid_sc)
r2_score_poly_sc = r2_score(y_test,y_pred_poly_scc)
```

```python
print(f"linear_r2score={r2_score_sc},rbf_r2score={r2_score_rbf_sc},sigmoid_r2score={r2_score_sigmoid_sc},poly_r2score={r2_score_poly_sc}")
```

```python
#Adding the "C" parameter  in model to improve the performance
from sklearn.svm import SVR
regressor_linear_c=SVR(kernel="linear",C=100)
regressor_rbf_c=SVR(kernel="rbf",C=100)
regressor_sigmod_c=SVR(kernel="sigmoid",C=100)
regressor_poly_c=SVR(kernel="poly",C=100)


```

```python
regressor_sigmod_c.fit(x_train_sc,y_train)
regressor_poly_c.fit(x_train_sc,y_train)
regressor_rbf_c.fit(x_train_sc,y_train)
regressor_linear_c.fit(x_train_sc,y_train)
```

```python
y_pred_linear_sc_c=regressor_linear_c.predict(x_test_sc)
y_pred_rbf_sc_c=regressor_rbf_c.predict(x_test_sc)
y_pred_sigmoid_sc_c=regressor_sigmod_c.predict(x_test_sc)
y_pred_poly_sc_c=regressor_poly_c.predict(x_test_sc)
```

```python
from sklearn.metrics import r2_score
r2_score_sc = r2_score(y_test,y_pred_linear_sc_c)
r2_score_rbf_sc = r2_score(y_test,y_pred_rbf_sc_c)
r2_score_sigmoid_sc = r2_score(y_test,y_pred_sigmoid_sc_c)
r2_score_poly_sc = r2_score(y_test,y_pred_poly_sc_c)
print(f"linear_r2score={r2_score_sc},rbf_r2score={r2_score_rbf_sc},sigmoid_r2score={r2_score_sigmoid_sc},poly_r2score={r2_score_poly_sc}")
```
