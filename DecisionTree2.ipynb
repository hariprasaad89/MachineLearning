```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading the file & Applying the one-hot encoding for the categorical columns
dataset=pd.read_csv("50_Startups.csv")
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
dataset
```

```python
#Getting columns details about dataset
dataset.columns
```

```python
#splitting input and output
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent=dataset[['Profit']]
```

```python
#splitting into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Creating the model
from sklearn.tree import DecisionTreeRegressor
decregressor= DecisionTreeRegressor()
decregressor=decregressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(decregressor)
plt.show
```

```python
y_pred=decregressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
#building the package
import pickle as pk
filename="decision_model.sav"
pk.dump(decregressor,open(filename,'wb'))
```
